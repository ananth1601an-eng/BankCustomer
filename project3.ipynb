```python
import pandas as pd
import numpy as np
import pickle
```

```python
df=pd.read_csv("bank_churn_dataset.csv")
df.head()
```

```python
df.tail()
```

```python
df = df.dropna()
print(df.shape)
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.describe(include='all')
```

```python
print(df.dtypes)
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str)
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.lower()
```

```python
df.sample(5)
```

```python
with open("bank_churn_dataset.pkl", "wb") as file:
    pickle.dump(df, file)

print("Pickle file saved successfully.")
```

```python
with open("bank_churn_dataset.pkl", "rb") as file:
    data = pickle.load(file)
```
